```python
# creating a patern for regular expression
pattern ="\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s\S{2}\s-\s"  #string of regex
```

```python
message=re.split(pattern,data)[1:] #selecting messages
```

```python
# selecting dates
dates=re.findall(pattern,data)
print(dates)
```
